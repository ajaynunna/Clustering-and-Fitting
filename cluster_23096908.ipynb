```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset loading
dat = pd.read_csv('D:\Finally_Done\Iris.csv')
df = pd.DataFrame(dat)
df
```

```python
# Target attributre
y = df['Species']
y
```

```python
# Plotting the histogram of ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

col_list = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for i in col_list:
    plt.hist(df[i])
    plt.title("The histogram of column "+i+" -- ")
    plt.show()

```

```python
# Plotting scatterplot between 'SepalLengthCm' and 'SepalWidthCm'
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'])
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("Scatterplot between 'SepalLengthCm' and 'SepalWidthCm'")
plt.show()

# Plotting scatterplot between 'PetalLengthCm' and 'PetalWidthCm'
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'])
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.title("Scatterplot between 'PetalLengthCm' and 'PetalWidthCm'")
plt.show()
```

```python
# Plotting the boxplots of ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i in col_list:
    plt.boxplot(df[i])
    plt.title("The boxplot of column "+i+" -- ")
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
ms = MinMaxScaler()
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
X = ms.fit_transform(X)
print("The standardized version of features -- ")
X
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay

y = pd.factorize(y)[0]
wcss = []
for i in range(1, 11):
    initlize = []
    for j in range(i):
      initlize.append(X[j])
    initlize = np.asarray(initlize)
    kmeans = KMeans(n_clusters=i, init=initlize, max_iter=100, n_init=1,random_state= 30)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    y_pred = kmeans.fit_predict(X)
    for m in range(0,len(y_pred)):
      if(y_pred[m]==2):
        y_pred[m]=0
      elif(y_pred[m]==0):
        y_pred[m]=2

    print("The accuracy for k =",i,"is =",accuracy_score(y_pred,y)," and the inertia value =",kmeans.inertia_)
    
    
# Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.title("Elbow Curve")
plt.show()
```

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
initlize = []
for j in range(3):
  initlize.append(X[j])
initlize = np.asarray(initlize)
kmeans = KMeans(n_clusters=3, init=initlize, max_iter=100, n_init=1,random_state= 30)
y_kmeans = kmeans.fit_predict(X)
for m in range(0,len(y_kmeans)):
  if(y_kmeans[m]==2):
    y_kmeans[m]=0
  elif(y_kmeans[m]==0):
    y_kmeans[m]=2

print("The predicted category/cluster values are = \n",y_kmeans)
print("Actual category/cluster values are = \n",y)
```

```python
print("Model evaluation for Iris data -- \n")
print("The accuracy for k =",3,"is =",accuracy_score(y_kmeans,y))
cm = confusion_matrix(y_kmeans, y) 
print ("\nConfusion Matrix : \n", cm)
print(classification_report(y_kmeans, y))
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.title("Confusion Matrix")
plt.show()
```

```python
# Visualising the clusters
fig = plt.figure(figsize =(10, 7))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 90, c = 'yellow', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 90, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 90, c = 'red', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroids')
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.title("Scatterplot between 'PetalLengthCm' and 'PetalWidthCm'")
plt.legend()
plt.show()
```
